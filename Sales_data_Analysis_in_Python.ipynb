```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Cargando los datos desde el dataframe
data = pd.read_csv('/content/Sales_Data.csv', index_col=0, encoding='windows-1252')

```

```python
data.head(10)
```

```python
#Tratamiento de datos que faltan
data.info()
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
#Eliminamos esos puntos en total 12 para que no causen perturbaciones en las graficas
data.dropna(axis=0, inplace=True)
data.isnull().sum()
```

```python
data.head()
```

```python
# Define el color de cada barra
palette = sns.color_palette("husl", len(data['Marital_Status'].unique()))
sns.countplot(x='Marital_Status', data=data, palette=palette)
plt.xlabel('Marital_Status')
plt.ylabel('count')
plt.show()
```

```python
# Define el color de cada barra
palette = sns.color_palette("husl", len(data['Gender'].unique()))
sns.countplot(x='Gender', data=data, palette=palette)
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()
```

```python
palette = sns.color_palette("husl", len(data['Zone'].unique()))
sns.countplot(x='Zone', data=data, palette=palette)
plt.xlabel('Zone')
plt.ylabel('count')
plt.show()
```

```python
palette = sns.color_palette("husl", len(data['Product_Category'].unique()))
plt.figure(figsize=(15,10))

plt.xticks(rotation=90)

sns.countplot(x='Product_Category', data=data, palette=palette)
```

```python
palette = sns.color_palette("husl", len(data['Occupation'].unique()))
plt.figure(figsize=(15,10))

plt.xticks(rotation=90)

sns.countplot(x='Occupation', data=data, palette=palette)
```

```python
#Ocupacion por sexo
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot(x='Occupation', hue='Gender', data=data)
plt.show()
```

**ClasificaciÃ³n de las ventas en funciÃ³n de la categorÃ­a del producto y el sexo**

```python
#Categoria de Producto por sexo
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot(x='Product_Category', hue='Gender', data=data)
plt.show()
```

**Analizando las ventas por grupo de edad y ocupaciÃ³n**

```python
#Ocupacion y rango de edades.
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot(x='Occupation', hue='Age Group', data=data)
plt.show()
```

El grÃ¡fico muestra que:
*   La barra azul que representa al grupo de edad de 26 a 35 aÃ±os es mÃ¡s permanente en cada una de las ocupaciones.
*   La barra naranja que representa a los jÃ³venes de 0 a 17 aÃ±os no compra nada.
*   Los mayores de 55 aÃ±os no compran tanto, pero en comparaciÃ³n con otras categorÃ­as, la barra es mÃ¡s prominente en la categorÃ­a de Sanidad (quizÃ¡s mÃ©dicos).
*  Las ventas totales se basan Ãºnicamente en el grupo de edad de 26-35 aÃ±os. El grupo de 35 a 40 aÃ±os tambiÃ©n es activo, pero no tanto.
